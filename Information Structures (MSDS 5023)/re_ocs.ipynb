```python
import re
import itertools
```

```python
#open and read the file
with open('/Users/osamples/sols/nightfive/samplehl7', 'r') as f:
    hl7 = f.read()
```

```python
#create a function for splitting the segments using regular expression
def get_segs(txt):
    re1 = re.compile("\n")    
    return re1.split(txt)

##works for now, but if i were to iterate through multiple strings it would not work
#instead keep a seperate variable for each as follows:
        #pattern = re.compile("\n")
        #split = re.split(pattern, split)
```

```python
#call the function by passing in the file
segs = get_segs(hl7)
segs
```

```python
#keep the for loops outside of the function. make the function as simple as possible
def get_fields(txt):
    pat = re.compile("\|")
    fields = re.split(pat, txt)
    return fields
```

```python
#for each segment in segs, pass it into get_fields()
fields = [get_fields(seg) for seg in segs]
fields
```

```python
#for field in fields:
    #print (len(field))
#they are not going to be the same length like you might have thought
```

```python
#uses itertools to take out the list of lists
#the astrix * means please unpack this list one thing at a time and then passes it into the calling function
#the calling function "chain" is like append but does not need existing list
#we could switch list() out for dict() and make everything a dictionary --there are other things it could be switched for too
merged_fields = list(itertools.chain(*fields))
```

```python
def get_comp(txt):
    pat = re.compile("\^")
    comps = (re.split(pat, txt))        
    return comps
```

```python
#for each merged field in merged_fields, pass it into get_comp
#this is called list comprehension (there is also dictionary comprehension and set comprehension)
comps = [get_comp(mf) for mf in merged_fields]
```

```python
#flatten this again so that it is just one list
merged_comps = list(itertools.chain(*comps))
```
