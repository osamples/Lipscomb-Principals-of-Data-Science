### Q2e. [10pts] What was the average house price in 2013?

````python
# YOUR CODE HERE
tt = df[df['Year'] == 2013]
sum = tt['Sale Price'].sum()
count = tt['Year'].count()
avg = sum/count
print("The average house price in 2013 was $%.2f" % avg)
````

### Q2f. [10pts] What was the average price of 4-bedroom houses in the Nashville area between 2013 and 2016?

````python
# YOUR CODE HERE
nash = df[df['Property City'].str.upper() == 'NASHVILLE']
four_nash = nash[nash['Bedrooms'] == 4]
sum = four_nash['Sale Price'].sum()
count = four_nash['Year'].count()
avg = sum/count
print("The average price of a 4-bedroom house in the Nashville area between 2013 and 2016 was $%.2f." % avg )
````
